```python
import pandas as pd
```

```python
df = pd.read_csv('211210-DRM-total.csv')
metal_df = df.iloc[:,2:25]
```

```python
metal_list = []
for i in range(df.shape[0]):
    metal_row = metal_df.iloc[i:i+1,:]
    metal_list.append(metal_row)
```

```python
metal_list[2]
```

```python
zero_counts = (metal_df == 0).sum(axis=1)
zero_counts = zero_counts.value_counts()
```

```python
import matplotlib.pyplot as plt

total_features = metal_df.shape[0]

plt.figure(figsize=(12, 9))

bars = plt.bar(
    zero_counts.index,
    zero_counts,
    color='steelblue',
    edgecolor='black',
    linewidth=0.6
)

plt.title('Number of Zero Values in Metal Features', fontsize=16,fontweight='bold')
plt.xlabel('The Amount of Zeros', fontsize=12)
plt.ylabel('The Amount', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.5)

# ìˆ«ìž + í¼ì„¼íŠ¸ í‘œì‹œ
for idx, count in zip(zero_counts.index, zero_counts):
    percent = count / total_features * 100
    plt.text(
        idx,
        count + 0.3,                      # ë§‰ëŒ€ ìœ„ ì‚´ì§ ë„ì›€
        f'{count}\n({percent:.1f}%)',     # ì¤„ë°”ê¿ˆìœ¼ë¡œ ì •ë¦¬
        ha='center',
        va='bottom',
        fontweight = 'bold',
        fontsize=12
    )
   # í–‰ì´ ë§Žì„ ê²½ìš° ì •ì„

plt.tight_layout()
plt.xticks(zero_counts.index, fontweight='bold')
plt.show()
```
